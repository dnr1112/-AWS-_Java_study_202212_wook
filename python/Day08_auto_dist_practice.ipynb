```python
import numpy as np
import math

def dist_np(p1, p2):
    return math.sqrt(sum((p2 - p1)**2))
```

```python
csv_data = []
with open('전국자동차정비업체표준데이터.csv', encoding='cp949') as f:
    for line in f:
        csv_data.append(line.split(','))
print(csv_data[:3])
```

```python
lat_list = []
long_list = []
for i in csv_data[1:]:
    lat_list.append(i[1])
    long_list.append(i[2])
print(lat_list)
print(long_list)
```

```python
new_list1 = []
new_list2 = []
for i in lat_list:
    try:
        new_list1.append(float(i))
    except:
        pass
lat_list = new_list1

for i in long_list:
    try:
        new_list2.append(float(i))
    except:
        pass
long_list = new_list2

lat_np = np.array(lat_list)
long_np = np.array(long_list)
print(lat_np)
print(long_np)
```

```python
dict_locate = dict(zip(lat_np, long_np))
print(list(dict_locate.items()))
```
