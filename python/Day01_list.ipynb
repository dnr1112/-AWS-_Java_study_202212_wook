```python
list_a = [i for i in range(10)]
list_b = [i for i in range(100, 110)]
print(list_a)

#in - T/F 반환
if 8 in list_a:  # <-> not in
    print('True')
else:
    print('False')
    
# concat
list_o = list_a + list_b
print(list_o)

# *
print(list_a * 2)
```

```python
list_a = [1,2,3]
list_b = list('abc')
d1 = {'a' : 1, 'b' : 2, 'c' : 3}
print(list_a)
print(list_b)
print(d1)

d2 = dict(zip(list_b, list_a))
print(d2)
print(d2.items())

d3 = dict([('a', 1), ('b', 2), ('c', 3)])
print(d3)

print(d3['c'])
print(d3.get('c'))
print(d3.get('x'))
d3['y'] = 15
print(d3)
```

```python
for i in d3:
    print(i, d3[i])
    
print(d3.keys())
print(d3.values())

for i in d3.values():
    print(i)
    
print(d3)
d3['y'] = 20
print(d3)
```
