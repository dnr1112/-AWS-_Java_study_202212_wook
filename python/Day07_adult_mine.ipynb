### 상관도 분석: UCK Repo Adult 데이터

* url : https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data

* age / capital gain 상관도

* hour-per-week / capital gain 상관도

```python
import requests

x = requests.get('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data')
```

```python
import math

def mean(num_list):
    for i in num_list:
        avr = sum(num_list) / len(num_list)
    return avr

def median(num_list):
    n = len(num_list)
    if n % 2 == 0:
        return (num_list[n//2 - 1] + num_list[n//2]) / 2
    else:
        return num_list[n//2]
        
def dev(num_list):    # 편차
    m = mean(num_list)
    return [x - m for x in num_list]

def ver(num_list):    # 분산
    n = len(num_list)
    d = dev(num_list)
    return sum([x * x for x in d]) / (n - 1)
    
def stdev(num_list):  # 표준편차
    return math.sqrt(ver(num_list))

def covar(list_1, list_2):  # 공분산
    n = len(list_1)
    list_1_dev = dev(list_1)
    list_2_dev = dev(list_2)
    return sum(x * y for x, y in zip(list_1_dev, list_2_dev)) / (n - 1)

def corel(list_1, list_2):  # 상관도 (-1 ~ 1)
    return covar(list_1, list_2) / (stdev(list_1) * stdev(list_2))
```

```python
pass1 = x.text.split('\n')

pass2 = []
for e in pass1:
    y = e.split(', ')
    pass2.append(y)
print(pass2[:10][0])
```

```python
age_list = []
for i in pass2:
    age_list.append(i[0])

print(age_list)
```

```python
new_list1 = []
for i in age_list:
    try:
        new_list1.append(float(i))
    except:
        pass
age_list = new_list1
print(age_list)
```

```python
capital_list = []
for i in pass2:
    try:
        capital_list.append(i[10])
    except:
        pass
print(capital_list)
```

```python
new_list2 = []
for i in capital_list:
    try:
        new_list2.append(float(i))
    except:
        pass
capital_list = new_list2
print(capital_list)
```

```python
hours_per_week_list = []
for i in pass2:
    try:
        hours_per_week_list.append(i[12])
    except:
        pass
print(hours_per_week_list)
```

```python
new_list3 = []
for i in hours_per_week_list:
    try:
        new_list3.append(float(i))
    except:
        pass
hours_per_week_list = new_list3
print(hours_per_week_list)
```

```python
print('상관도: ', corel(age_list, capital_list))
print('상관도: ', corel(hours_per_week_list, capital_list))
```
